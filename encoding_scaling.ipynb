```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, RobustScaler

# ── Original Dataset ───────────────────────────────────────────────
data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City':   ['Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Size':   ['Small', 'Large', 'Medium', 'Large', 'Small'],
    'Age':    [25, 45, 32, 60, 28],
    'Fare':   [15, 300, 85, 450, 20]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# ══════════════════════════════════════════════════════════════════
# TASK 1 — Categorical Encoding
# ══════════════════════════════════════════════════════════════════

# ── 1a. Gender → Label Encoding ───────────────────────────────────
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
# Female=0, Male=1  (alphabetical order by default)
print("\nAfter Label Encoding Gender:")
print(df[['Gender']])

# ── 1b. City → One-Hot Encoding (drop_first=True avoids dummy trap)
df = pd.get_dummies(df, columns=['City'], drop_first=True)
print("\nAfter One-Hot Encoding City:")
print(df)

# ── 1c. Size → Ordinal Label Encoding (Small=0, Medium=1, Large=2)
size_order = {'Small': 0, 'Medium': 1, 'Large': 2}
df['Size'] = df['Size'].map(size_order)
print("\nAfter Ordinal Encoding Size:")
print(df[['Size']])

# ══════════════════════════════════════════════════════════════════
# TASK 2 — Feature Scaling
# ══════════════════════════════════════════════════════════════════

# RobustScaler is chosen here because Fare contains an outlier (450).
# Unlike StandardScaler (which uses mean) or MinMaxScaler (which uses min/max),
# RobustScaler uses the median and IQR, so outliers do not distort the scaling.

scaler = RobustScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])

print("\nAfter RobustScaler on Age and Fare:")
print(df)
```
